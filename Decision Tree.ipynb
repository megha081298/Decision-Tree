```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df["species"].value_counts()
```

```python
sns.pairplot(data=df,hue="species")
```

```python
x=df.select_dtypes(include=["float64","int64"])
x=x.corr(min_periods=1)
sns.heatmap(x.corr(),cmap="Blues",annot=True)
```

```python
x=df.iloc[:,0:4].values
y=df.iloc[:,-1]
```

```python
y
```

```python
sns.get_dataset_names()
```

```python
y.value_counts()
```

```python
#Splitting the data:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=89)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm,annot=True,cmap="Blues")
plt.xlabel("Predicted values")
plt.ylabel("Actual Values")
```

```python
plt.figure(figsize=(10,15))
from sklearn.tree import plot_tree
plot_tree(decision_tree=dtree,feature_names=df.columns,
         class_names=["setosa","versicolor","virginica"],filled=True,precision=4,rounded=True)
```

# Hyperparameter Tuning:

```python
from sklearn.model_selection import GridSearchCV

criteria=["gini","entropy"]
min_sample_split_range=[2,3,10,20,67]
max_depth_range=[2,3,5,10]
min_sample_leaf=[1,5,10]
max_leaf_nodes=[5,10,20,15]

param_grid={"criterion":criteria,"min_samples_split":min_sample_split_range,
           "max_depth":max_depth_range,"min_samples_leaf":min_sample_leaf,
           "max_leaf_nodes":max_leaf_nodes}

grid=GridSearchCV(estimator=dtree,param_grid=param_grid,cv=5,scoring="accuracy",refit=True)


grid.fit(x_train,y_train)
print(grid.best_params_)
print(grid.best_score_)
```
